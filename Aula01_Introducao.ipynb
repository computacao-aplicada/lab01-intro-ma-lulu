## O que √© Vis√£o Computacional?

A **Vis√£o Computacional** √© uma √°rea da Intelig√™ncia Artificial que busca permitir que os computadores 'vejam' e interpretem o mundo visual, assim como os humanos.

Ela utiliza imagens e v√≠deos como entrada e tenta extrair informa√ß√µes √∫teis, como:
- Detec√ß√£o de objetos
- Reconhecimento facial
- Segmenta√ß√£o de imagem
- Rastreamento de movimento

A **Computa√ß√£o Gr√°fica**, por outro lado, est√° mais voltada √† gera√ß√£o de imagens ‚Äî como renderizar cenas em 3D.

J√° o **Processamento de Imagens** se concentra na transforma√ß√£o e an√°lise de imagens (normalmente 2D) para melhorar a qualidade ou extrair caracter√≠sticas.

| √Årea                  | Foco                             |
|-----------------------|----------------------------------|
| Vis√£o Computacional   | Entendimento e interpreta√ß√£o     |
| Computa√ß√£o Gr√°fica    | Gera√ß√£o de imagens               |
| Processamento de Imagens | Manipula√ß√£o e melhoria de imagens |


```python
# Importa√ß√£o das bibliotecas necess√°rias
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Exibir gr√°ficos inline
%matplotlib inline
```

## Exemplo pr√°tico: Carregar e visualizar uma imagem

```python
# Carregar imagem de exemplo
from PIL import Image
import requests

url ="https://images.unsplash.com/photo-1742201949705-220268ad4467?q=80&w=1528&auto=format&fit=crop&ixlib=rb-4.1.0&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D"
image = Image.open(requests.get(url, stream=True).raw)
image = image.convert("RGB")  # Garante 3 canais

# Mostrar imagem
plt.imshow(image)
plt.title("Imagem Original")
plt.axis('off')
plt.show()
```

## Manipula√ß√£o dos canais RGB

```python
# Separar os canais
r, g, b = image.split()

# Exibir os canais
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
for ax, channel, color in zip(axs, [r, g, b], ['R', 'G', 'B']):
    ax.imshow(channel, cmap='gray')
    ax.set_title(f'Canal {color}')
    ax.axis('off')
plt.tight_layout()
plt.show()
```

## Convers√£o para escala de cinza

```python
# Converter para escala de cinza
gray_image = image.convert("L")

# Mostrar imagem em tons de cinza
plt.imshow(gray_image, cmap='gray')
plt.title("Imagem em Escala de Cinza")
plt.axis('off')
plt.show()
```

## Explorando os pixels

```python
# Converter para numpy e mostrar valor de alguns pixels
np_image = np.array(image)
print("Formato da imagem:", np_image.shape)
print("Valor do pixel (0, 0):", np_image[0, 0])
```

##### Exerc√≠cio 1

```python
# Ex.1 Exibir apenas o canal vermelho da imagem.

# Separar os canais
r, g, b = image.split()

# Exibir os canais
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
for ax, channel, color in zip(axs, [r, g, b], ['R', 'G', 'B']):
    ax.imshow(channel, cmap='Reds')
    ax.set_title(f'Canal {color}')
    ax.axis('off')
plt.tight_layout()
plt.show()
```

##### Exerc√≠cio 2

```python
# Ex.2 Modifique aleatoriamente o canal vermelho e gere a imagem novamente.

np_image = np.array(image)
np_image[:, :, 0] = np.clip(np_image[:, :, 0] * 1.5, 50, 255)  # Aumenta R

# Exibir
plt.figure(figsize=(12, 6))
plt.subplot(121); plt.imshow(image); plt.title("Original")
plt.subplot(122); plt.imshow(np_image); plt.title("Canal R modificado")
plt.show()
```

##### Exerc√≠cio 3

```python
# Ex.3 Criar uma imagem negativa (inverter as cores).

img_array = np.array(image)

# Inverter todos os canais (R, G, B)
img_invertida = 255 - img_array

# Converter de volta para imagem
img_final = Image.fromarray(img_invertida)

# Exibir compara√ß√£o
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].imshow(image)
axs[0].set_title("Original")
axs[1].imshow(img_final)
axs[1].set_title("Negativo da Imagem")
plt.show()
```

```python
# Ex.5 Recorte um peda√ßo da imagem e salve em disco.
from PIL import Image
import requests
from IPython.display import display, FileLink

url = "https://images.unsplash.com/photo-1742201949705-220268ad4467?q=80&w=1528&auto=format&fit=crop&ixlib=rb-4.1.0&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D"  # Insira a URL da imagem aqui
image = Image.open(requests.get(url, stream=True).raw)

# 2. Definir a regi√£o de recorte (esquerda, superior, direita, inferior)
left = (800 - 400) // 2
top = (600 - 400) // 2
right = left + 400
bottom = top + 400
regiao_recorte = (left, top, right, bottom)

# 3. Recortar a imagem
imagem_recortada = image.crop(regiao_recorte)

# 4. Salvar o recorte
imagem_recortada.save("recorte.jpg")

# 5. Resultado
display(imagem_recortada)
```
